# Seaborn: regplot


# Types of Regressions
- order,robust , logistical,lowess
- DEFAULT - Linear regression (order=1)
- ORDER - polynomial regression
- ROBUST - robust regression exclude outliers
- LOGISTIC - logistic regression, binary classification
- LOWESS - locally weighted regression, smooths out data


```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds.sample(n=200, random_state=44)

diamonds.shape
```

## Intro Visuals

```python
import numpy as np
```

```python
sns.set_style('white')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
blue, orange, green, red = sns.color_palette()[:4]
```

```python
x_vals = np.random.rand(100)*5
y_vals_posTwo = x_vals*2 + np.random.rand(100)*4
y_vals_posHalf = x_vals*0.5 + np.random.rand(100)*3 + 2
y_vals_negOne = x_vals*(-1) + np.random.rand(100)*3 + 6
```

```python
plt.figure(figsize=(3, 5))
sns.regplot(x_vals, y_vals_posTwo, scatter_kws={'alpha': 0.4}, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
sns.regplot(x_vals, y_vals_posHalf, scatter_kws={'alpha': 0.4}, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
sns.regplot(x_vals, y_vals_negOne, scatter_kws={'alpha': 0.4}, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_posTwo, scatter=False, ci=None, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_posHalf, scatter=False, ci=None, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_negOne, scatter=False, ci=None, line_kws={'lw': 4})
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_posHalf, fit_reg=False)
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_posTwo, fit_reg=False)
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

```python
plt.figure(figsize=(3, 5))
plt.gca().set(xlim=(0, 5))
sns.regplot(x_vals, y_vals_negOne, fit_reg=False)
sns.despine()
plt.xlim(0, 5)
plt.ylim(0, 12)
plt.tight_layout();
```

## Basics

```python
sns.regplot(diamonds.carat, diamonds.price);
```

```python
sns.regplot(x='carat', y='price', data=diamonds);
```

```python
sns.regplot(x='carat', y='price', data=diamonds, fit_reg=False); # fit_reg=False removes the regression line
```

```python
plt.gca().set(xlim=(0,2.6))
sns.regplot(x='carat', y='price', data=diamonds, scatter=False); # Switch off the scatter plot
```

## regplot Options

### Confidence Intervals

```python
sns.regplot(x='carat', y='price', data=diamonds,
            ci =95, # 95% confidence interval
            );
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
            ci = None, # Switch off the confidence interval
            );
```

### Discrete Variables

```python
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}

diamonds['cut_value'] = diamonds.cut.map(cut_map)
diamonds['cut_value'] = diamonds.cut_value.astype('category')
```

```python
diamonds.cut_value.value_counts()
```

```python
diamonds.head()
```

### Models

#### Polynomial Regression (order)

```python
sns.regplot(x='carat', y='price', data=diamonds,
            fit_reg=False
               );
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
            order=2
            );
```

#### Robust Regression

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None
            );
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None,
            robust=True # Robustly fit a regression model
            );
```

## Related Seaborn Plots

```python
blue, orange, green, red = sns.color_palette()[:4]
```

```python
p = sns.lmplot(x='carat', y='price', 
               data=diamonds[diamonds.color.isin(['E', 'J'])], 
               hue='color', 
               order=2,
               palette=[green, orange])
plt.xlabel('Carat', fontsize=18)
plt.ylabel('Price', fontsize=18)
p._legend.remove()
plt.legend(fontsize=16)
plt.xticks([])
plt.yticks([])
plt.tight_layout();
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, 
              kind='reg', 
              color='purple')
plt.xlabel('')
plt.ylabel('')
plt.xlim(-0.1, None)
plt.ylim(-2000, None)
plt.tight_layout();
```

```python
sns.pairplot(diamonds[['carat', 'depth', 'price']], 
             kind='reg', 
             palette='colorblind')
plt.tight_layout();
```
