# Seaborn: lmplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris').dropna()

df.head()
```

## Basics

```python
sns.set_style('white')
```

```python
sns.lmplot(x='sepal_width', y='sepal_length', data=df)
```

### `hue`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df)
```

### `markers`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df,markers=['o','s','D'])
```

## Small Multiples (FacetGrid)

### `col`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df, col='species', palette='Set1')
```

### `row`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df, row='species', palette='Set1',truncate=True)
```

## Extras

### `x_bins`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df[df['species']=='versicolor'], row='species', col='species', palette='Set1', truncate=True)
```

### Other models: Logistic Regression, `logistic`

## Styling

### `scatter_kws`

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=df, row='species', col='species', palette='Set1', truncate=True,scatter_kws={'s':100})
```

### `line_kws`

```python
sns.lmplot(x='culmen_depth_mm', y='culmen_length_mm', 
           data=df, 
           hue='species',
           col='species',
           line_kws={'lw': 5, 'ls': '-.', 'color': 'black'}
           );
```

### FacetGrid Object

```python
g = sns.lmplot(x='culmen_depth_mm', y='culmen_length_mm', 
               data=df, 
               hue='species', 
               col='species'
              );
```

```python
type(g)
```

```python
g = sns.lmplot(x='culmen_depth_mm', y='culmen_length_mm', 
               data=df, 
               hue='species', 
               col='species'
              )

g.set_titles(col_template='{col_name} Penguins')
g.set_xlabels('Culmen Depth (mm)');
```

## Examples, Thumbnail

```python
from matplotlib import pyplot as plt
```

```python
sns.set(style='white', font_scale=1.7)
g = sns.lmplot(x='culmen_length_mm', y='culmen_depth_mm', data=df, hue='species', col='species', truncate=False)

g.set_titles('{col_name} Penguins')
g.set_xlabels('Culmen Length (mm)')
g.set_ylabels('Culmen Depth (mm)')
plt.tight_layout()
#plt.savefig('20lmplot_image_facetGridExample.png', dpi=200);
```

```python
diamonds = sns.load_dataset('diamonds').sample(1000)
```

```python
sns.lmplot(x='carat', y='price', data=diamonds, hue='clarity', legend=False)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.ylim(0, None)
plt.xlim(0, None)
plt.legend(loc=2, title='clarity')
plt.tight_layout()
#plt.savefig('20lmplot_image_tooManyLines.png', dpi=200);
```

```python
sns.set(style='white', font_scale=2)
sns.lmplot(x='carat', y='price', hue='clarity', data=diamonds, col='clarity', col_wrap=3)
plt.xlim(0, None)
plt.ylim(0, None)
#plt.savefig('20lmplot_image_tooManyLinesFacet.png', dpi=200);
```
