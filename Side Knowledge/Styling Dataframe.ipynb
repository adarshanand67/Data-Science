````python
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
````

1. Formatting numbers


````python
format_dict = {
    'total_bill': '${:,.2f}',
    'tip': '${:,.2f}',
}
````

````python
df.head().style.format(format_dict)
````

2. Hiding index

````python
df.head().style.hide_index()
````

3. highlight_max

````python
df.head().style.highlight_max(color='red')
````

4. highlight_min

````python
df.head().style.highlight_min(color='green')
````

````python
df.head().style.highlight_min('size',color='green') # Highlight the minimum size a column 
````

5. Background_gradient

````python
df.head().style.background_gradient(cmap='RdYlGn_r')
````

````python
df.head().style.background_gradient(cmap='RdYlGn_r', subset=['tip']) # Gradient color for tip column
````

6. Bar

````python
df.head().style.bar() 
````

````python
df.head().style.bar(subset=['tip']) # Bar chart for tip column
````

7. set_caption

````python
df.head().style.set_caption("Tips Dataframe")
````

8. setting properties (center all the text)

````python
df.head().style.set_properties(**{'text-align': 'center'})
````

````python
df.head().style.set_properties(**{'width': '100px'})
````

eg - Final example of aesthetic dataframes

````python
df.head().style.set_caption("This is Adarsh anand").set_precision(2).background_gradient().hide_index()
````
