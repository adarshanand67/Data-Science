## Most efficient way to create a table from wiki data
## (Fastest web scraping)

```python
import pandas as pd
import wikipedia as wp
```

```python
def getTableWIKI(content,table_number=0):
    ''' Get table from wikipedia page '''
    import wikipedia as wp
    import pandas as pd
    
    html = wp.page(content).html().encode("UTF-8")
    
    try: 
        df = pd.read_html(html)[table_number]  # Try 2nd table first as most pages contain contents table first
    except IndexError:
        print('No table found')
        
    return df

df = getTableWIKI('List of S&P 500 companies')
df.head()
```

```python
tickers = df.Symbol.to_numpy()
tickers
```

```python
df = getTableWIKI("Android version history")
df
```
