```python
data_ECB = r"""Towards Data Science
Sign in

Get started
Follow
614K Followers
·
Editors' Picks
Features
Deep Dives
Grow
Contribute
About




Guide to Using Word Clouds for Applied Research Design
Discover the essential applications for textual data visualization in applied research and use them for your project
Petr Korab
Petr Korab

Dec 14, 2021·6 min read







Image by authors
By: Petr Koráb (Lentiamo, Prague; Zeppelin University, Germany), Jarko Fidrmuc (Zeppelin University, Germany), and David Štrba (Lentiamo, Prague)
Introduction
Textual data, in various forms, including statements, speech, public announcements, and their effects, have been frequently analyzed in applied economic research. Since text can be quantified and used for modeling and predictions, researchers used various techniques to drill it and use it in their empirical design.
To be more specific, the word cloud is a simple yet widely used plot to display and understand the structure of textual data. All main statistical programs such as R, SPSS, Python, STATA, MATLAB, and Business Intelligence tools such as Tableau, Looker, and Power BI include a word cloud implementation. Word clouds have also developed more attractive features that include various shapes and backgrounds.
This short guide will introduce you to the four main applications of the word cloud for designing applied research and data science projects, along with an implementation in Python. You will learn how to:
understand the structure of data using word clouds
select the essential features for the model
understand the context in textual data
use word clouds to make data story-telling
What is a Word Cloud ?
Word cloud is an image composed of words used in a particular text, in which the size of each word indicates its frequency or importance. It is a simple graph that is frequently used in business and academia. Technically, a word cloud is based on n-grams that are in computational linguistics and probability fields sequences of n items from a sample of text or speech. The items can be syllables, letters, or words.
Using Latin numerical prefixes, an n-gram of size 1 is referred to as a “unigram”, size 2 is a “bigram”, and size 3 is a “trigram.” When N=1, unigrams are essentially the individual words in a sentence. For example, in the sentence “The cow jumps over the moon”, if N = 2, then the n-grams are: “the cow”, “cow jumps”, “jumps over”, “over the”, etc. If N = 3, the n-grams are:” the cow jumps”, “cow jumps over”, “jumps over the”, and “over the moon”.
N-grams are practically used in many data apps. One of them is the Google Books Ngram Viewer, an online search engine that charts the frequencies of any set of search strings using a yearly count of n-grams found in the digitalized repository of Google books.
In academia, researchers have recently used word clouds to display frequencies in textual data. In finance, Feldkircher and his colleagues collected speeches made by central bankers in Europe and analyzed their content. In the Journal of Economic Literature article, Bowles and Carlin use word clouds to illustrate the need for change in teaching economics. Already in 2018, Allen and McAleer explore Twitter data on global warming, climate change, and weather and use word clouds to plot the most prominent words in the tweets made by the former US president, Donald Trump.
Data summarization
Word clouds are designed to summarize textual data. Let’s demonstrate it on a speech made by Christine Lagarde, President of the European Central Bank (ECB), at the forum for central bankers in September 2021. The lecture is available on the ECB’s website. Even without reading, we can clearly understand the ECB governor’s topics at one glance. To achieve this, the key questions are:
What is the structure of the data?
What is the key topic in the data to focus on?
We will use Python’s matplotlib and wordcloud libraries:
The article text is copied to a string variable:
And, finally, we produce a word cloud with the summarized text.
With a couple of lines of code, we see that the ECB governor is mainly concerned about inflation, which is linked to pandemic, recovery, recession, wages, and supply. These are clearly the topics that the European economy faces after the two years of the covid crisis.

Image by authors
The benefits of this word cloud application dramatically increase with the data size. If the dataset contains tens or hundreds of MBs, then reading such text consumes a lot of time. On the other hand, running a python script takes a few seconds, can be automatized and provides the most relevant information for potential decisions.
Feature selection
Going beyond simple data summarization but following the same principle, we can select the most crucial factor and use it for further modeling. Feature selection is a term commonly used in machine learning, referring to choosing the essential variables for the model. Word clouds display word frequencies (or n-1 grams) or word co-occurrences (higher-order n-grams), and they essentially do the same with textual data.
The key questions are here:
What are the most important factors related to a specific topic?
The dataset for this example is the Coronavirus News Headline dataset, which is freely available at Kaggle. Here, more specifically, we can discuss the following question:
Which topics were mostly discussed during the Covid-19 pandemic?
We take only the first 2000 headlines from the dataset that contain the beginning of the Covid crisis. This subset is then the source data for the same python code as in the previous example.

Image by authors
As we can see, after the outbreak of Covid 19, the media focused on its origin: China, Wuhan, and its spread to other countries. Any social and economic research focusing on the beginning of the Covid crisis should reflect these crucial factors.
Animated word cloud
Word clouds are commonly presented as static plots without a time dimension. On the other hand, animated word cloud adds a dynamic feature into the analysis and helps tell stories about the data. Combining hundreds of pictures into an MP4 video, we can see the development of the significance of topics over time.
Michael Kane and his colleagues introduced this original approach, collected the 250 most frequent words found in the Science journal abstracts, and developed a dynamic version of word cloud called the Word Swarm. The Python codes are freely available from their GitHub.
We optimized the original code for Python 3.8. and used it to analyze titles of articles published in the five most prominent economic journals. Our data consist of word frequencies (i.e., unigrams) constructed from article titles from those journals. The animated word cloud shows the research trends in the data in an exciting video presentation. The article is available from here.
Source: Koráb, Štrba, Fidrmuc (2021). Animated Word Clouds: A Novel Way for the Visualization of Word Frequencies. In: Python in Plain English.
Text network visualization
Word clouds are used to generate visual summaries of textual data; however, they may lose track of the context. There is a risk of getting false results and making the wrong assumptions about the underlying data. Text network analysis solves this problem by considering the ‘words’ co-occurrences (i.e., higher-order n-grams) and suggests a different way of thinking about text: not a cloud of concepts but a cloud of relations between them.
InfraNodus is another tool that uses text network analysis as the core framework for text mining and textual data analysis. The text network approach can be summarized in the following steps:
import your data from various sources
generate a network structure
create a word cloud with relations to other related topics
Our earlier article provides a detailed introduction to InfraNodus, including an example insight into the google queries during the Covid-19 crisis. For more detailed information, also read the InfraNodus documentation and a word cloud case study.

Source: Koráb (2021). InfraNodus: An Excellent Tool for Textual Data Analysis. In: Towards Data Science.
Conclusions
Internet and other digital data sources provide vast quantities of textual information. For example, policy makers’ speeches are freely available on the web both for the public, decision-takers like investors and researchers. It has become really important to process and visualize this vast amount of information that standard methods cannot process (reading and manual summaries).
This article has suggested three applications of word clouds for applied research design in static, dynamic, and text network forms. We showed that these methods are very useful for designing the research in the early phase or presenting results in the later stages. They can also be used in various data science projects that use textual data, either in firms or elsewhere. The number of tools and their applications in various areas will certainly increase in the near future.
Sign up for The Variable
By Towards Data Science
Every Thursday, the Variable delivers the very best of Towards Data Science: from hands-on tutorials and cutting-edge research to original features you don't want to miss. Take a look.


Get this newsletter
Text Mining
NLP
Word Cloud
Python
Data Visualization
14







More from Towards Data Science
Follow
Your home for data science. A Medium publication sharing concepts, ideas and codes.

Read more from Towards Data Science
More From Medium
Can Data Science help us find what makes a hit television show?
Shilpi Bhattacharyya in Towards Data Science

The Million-Dollar model
Markus Proesch in The Sports Scientist

Using Machine Learning to Predict Home Prices.
shweta&data

The Five W’s of Evaluating an Organization’s Data Science Understanding
The Data Detective in Towards Data Science

HAMR@ISMIR2018 — Deezer HQ
Manuel Moussallam in Deezer I/O

GA COVID-19 Report July 11, 2020
Andrew Benesh

Unlimited Limitations
Chip

Student Exploration- Adding Vectors (answers)
Josh Brolin

About

Write

Help

Legal"""
```

```python
def Create_wordCloud(sample_text):
    """Create a word cloud from a text string."""
    
    # Importing necessary libraries
    from PIL import Image
    from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
    import matplotlib.pyplot as plt
    %matplotlib inline
    import numpy as np
    
    fig = plt.subplots(figsize=(24,20)) # set size
    stopwords = set(STOPWORDS) # set stopwords
    
    wordcloud = WordCloud(max_font_size=80, max_words=1000, background_color='white', colormap='hsv').generate(sample_text) # Generate word cloud
    
    wordcloud = WordCloud(background_color="white", max_words=1000, stopwords=stopwords,
               min_font_size=10, colormap='Dark2').generate(sample_text) # Generate word cloud
    
    # plt.title('Word cloud',fontsize=50) # set title
    plt.imshow(wordcloud) # show
    plt.axis('off') # no axis
    plt.show(); # show it
    
Create_wordCloud(data_ECB)
```
