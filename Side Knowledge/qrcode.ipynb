```python
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import *
from qrcode.image.styles.colormasks import *

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Basic Layout of QR Code
qr = qrcode.QRCode(version=1,error_correction=qrcode.constants.ERROR_CORRECT_L,box_size=10,border=4)

qr.add_data('Welcome to my Github account - https://github.com/AdarshAnand67')

qr.make(fit=True)
# qr = qr.make_image(fill_color="black", back_color="white")

type(qr)
qr
```

```python
img_2 = qr.make_image(image_factory=StyledPilImage, color_mask=SquareGradiantColorMask())
img_2
```

```python
img_3 = qr.make_image(image_factory=StyledPilImage)
img_3
```

```python
# Other module_drawers: RoundedModuleDrawer() , SquareModuleDrawer() , RectangularModuleDrawer(), DiagonalGradientModuleDrawer()
```

```python
img = qr.make_image(image_factory=StyledPilImage, module_drawer=RoundedModuleDrawer(), color_mask=RadialGradiantColorMask())
img
```

```python
# List of available module_drawers: RoundedModuleDrawer() , SquareModuleDrawer() , RectangularModuleDrawer(), DiagonalGradientModuleDrawer()
img = qr.make_image(back_color=(255, 195, 235), fill_color=(55, 95, 35))
img
```
