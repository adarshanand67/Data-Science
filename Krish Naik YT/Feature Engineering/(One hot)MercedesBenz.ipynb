```python
# TOP 10 categories of X2 column
top10 = [x for x in df['X2'].value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
# Create 10 new columns for each of the top 10 categories
for i in top10:
    df[i] = df['X2'].map(lambda x: 1 if x == i else 0)

df[['X2']+top10].head()
```

```python
def oneHotEncoding(df,cols):
    top10 = [x for x in df[cols].value_counts().sort_values(ascending=False).head(10).index]    

    for i in top10:
        df[i] = df[cols].map(lambda x: 1 if x == i else 0)
    return df

df = pd.read_csv('/home/adarshanand67/Documents/GitHub/Data-Science-Personal/Krish Naik/Feature Engineering/data/train.csv',usecols=['X1','X2','X3','X4','X5','X6'])

oneHotEncoding(df,'X2')
df
```

```python
# Advantage One hot encoding ? (Most frequent categories)

# Create 10 new columns for each of the top 10 categories 
# No need for variable transformation.
# Easily select varibles for training and testing just by 0,1

# Disadvantage
# Does not add any new information.
```
