```python
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("tips") # Inbuilt dataset tips from seaborn
```

```python
df.head()
# Tip - depedent feature (y)
# rest - independent feature(x)
```

### Correlation with Heatmap

 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

```python
'''
Correlation -
-1 to 1
-1 denotes negative correlation
1 denotes positive correlation
0 denotes no correlation
'''
df.corr() # only integer and float values are considered
```

```python
sns.heatmap(df.corr(),annot=True,center=True) # annot=True to show the values
```

## JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
### Univariate Analysis



```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex') # kind='hex' to show the hexagonal plot
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde') # kind='kde' to show the kernel density plot
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') # kind='reg' to show the regression plot
```

## Pair plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
df.head()
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.displot(df['tip']) #Distribution of the colums (Frequency count)
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
## Count plot

sns.countplot('smoker',data=df)
```

```python
df
```

```python
sns.countplot('sex',data=df)
```

```python
## Count plot

sns.countplot(x='day',data=df)
```

```python
## Bar plot - 
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
## Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.head()
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('day','total_bill', data=df,palette='rainbow')
```

```python
sns.boxplot(x="smoker", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```

```python
species = list(set(iris['species']))
species
import matplotlib.pyplot as plt
```

```python
# 2*2 subplots - distplot for virginica
plt.subplot(2,2,1)
sns.distplot(iris[iris['species']=='virginica']['sepal_length'])
plt.subplot(2,2,2)
sns.distplot(iris[iris['species']=='virginica']['sepal_width'])
plt.subplot(2,2,3)
sns.distplot(iris[iris['species']=='virginica']['petal_length'])
plt.subplot(2,2,4)
sns.distplot(iris[iris['species']=='virginica']['petal_width'])
```

```python
# 2*2 subplots - distplot for setosa
plt.subplot(2,2,1)
sns.distplot(iris[iris['species']=='setosa']['sepal_length'])
plt.subplot(2,2,2)
sns.distplot(iris[iris['species']=='setosa']['sepal_width'])
plt.subplot(2,2,3)
sns.distplot(iris[iris['species']=='setosa']['petal_length'])
plt.subplot(2,2,4)
sns.distplot(iris[iris['species']=='setosa']['petal_width'])
```

```python
# Category plot - show count of each species
sns.countplot(x='species',data=iris)
```

```python
# Box plot - show box plot for each species
plt.subplot(2,2,1)
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,2)
sns.boxplot(x='species',y='sepal_width',data=iris)
plt.subplot(2,2,3)
sns.boxplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,4)
sns.boxplot(x='species',y='petal_width',data=iris)

# increase size
plt.rcParams['figure.figsize'] = (16,16)
```
