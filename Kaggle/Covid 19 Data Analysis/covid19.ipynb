```python
# Death vs Confirmed in India
plt.figure(figsize=(10,8))
sns.scatterplot(x='Confirmed',y='Deaths',data=df,hue='WHO Region',palette='rainbow')
# Move the legend to the top left
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot()
```

```python
sns.pairplot(df,hue='WHO Region',palette='rainbow')
```

```python
# Deaths per 100 cases country wise
plt.figure(figsize=(10,10))
sns.set(style='whitegrid', palette='rainbow')
sns.scatterplot(x='Country/Region',y='Deaths / 100 Cases',data=df,hue='WHO Region')
plt.xticks(rotation=90)
# increase space between ticks
plt.subplots_adjust(bottom=0.15)
# remove top and right borders
plt.gca().spines["top"].set_visible(False)
# remove x labels
plt.gca().xaxis.set_tick_params(labelbottom=False)
# increase y ticks
plt.gca().yaxis.set_tick_params(labelsize=12)
# increase bubble size
plt.scatter(x=df['Country/Region'],y=df['Deaths / 100 Cases'],s=df['Deaths / 100 Cases']*100,alpha=0.6)
```

```python
print("Country with the most deaths per 100 cases:",df['Country/Region'][df['Deaths / 100 Cases'].idxmax()])
```
