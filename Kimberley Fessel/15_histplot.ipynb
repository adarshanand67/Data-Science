## Basics

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.flipper_length_mm);
```

```python
sns.histplot(y = 'flipper_length_mm', data=penguins);
```

### `kde`

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             kde=True
             );
```

### `bins`

```python
sns.histplot(x='flipper_length_mm', data=penguins, bins=20); 
```

### `binwidth` and `binrange`

```python
sns.histplot(x='flipper_length_mm', data=penguins, 
             binwidth=10, binrange=(180,250),
             kde=True
            ); 
```

## Statistics


### `stat`

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             stat='count'
            );
```

**density** - AREA of histogram sums to one

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             stat='density'
            );
# Probablity density function
```

**probability** - HEIGHT of histogram bars to one

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             stat='probability'
            );
# Probablity that the flipper length is greater in that space
```

### `cumulative`

```python
sns.histplot(x='flipper_length_mm', data=penguins,
             stat='probability', fill=False, element='step',
             cumulative=True
            );
```

## Categorical Variables

### `hue`

```python
sns.histplot(x='flipper_length_mm', data=penguins, hue='species');
```

### `element`

```python
sns.histplot(x='flipper_length_mm', data=penguins, 
             hue='species',
             element='step'
            );
```

```python
sns.histplot(x='flipper_length_mm', data=penguins, 
             hue='species',
             element='poly'
            );
# Creates polygons
```

### `multiple`

```python
sns.histplot(x='flipper_length_mm', data=penguins, 
             hue='species', 
             multiple='stack'
            ); 
```

```python
sns.histplot(x='flipper_length_mm', data=penguins, 
             hue='species', 
             multiple='fill'
            ); 

# Fill argument is used to fill the polygons , scaled upto 1.
```

**BONUS Example**: `multiple="dodge"`

```python
sns.histplot(x='species', data=penguins, 
             hue='island', 
             multiple='dodge',
             shrink=0.8
            ); 
```

## Bivariate Histplot

```python
sns.set_style('whitegrid')
```

```python
sns.histplot(x='flipper_length_mm', y='bill_length_mm', data=penguins);
```

### `cbar`

```python
sns.histplot(x='culmen_length_mm', y='culmen_depth_mm', data=penguins,
             cbar=True
            );
```

### `hue`

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
             hue='species',cbar=True,
            )
# Move legend out
plt.legend(loc='upper right');
```

### Discrete color map

```python
sns.histplot(x='culmen_length_mm', y='species', data=penguins,
             hue='species'
            );
```

## Styling 

### `shrink`

```python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins,
             shrink=0.8
            );
```

### `color`, `palette`

```python
sns.histplot(x='bill_depth_mm', data=penguins,
             color='pink'
            );
```

```python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins,shrink=0.8, palette='bone');
```

### `fill`

```python
sns.histplot(x='bill_depth_mm', data=penguins,
            fill=False
            );
            
```
